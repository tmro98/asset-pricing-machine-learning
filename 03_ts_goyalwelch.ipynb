```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import glob
from pathlib import Path
```

```python
df = pd.read_csv("data/other/GoyalWelchMonthly.csv")
```

```python
df = df.rename(columns={"yyyymm":"Date"})
df.index = pd.to_datetime(df["Date"], format="%Y%m")
df.drop(["Date", "csp"],axis=1, inplace=True)
df.dropna(axis=0, inplace=True)
df.index = df.index.to_period("M")
df = df.replace(',','', regex=True)
df = df.astype(float)
```

```python
df["Excess"] = df["Index"].pct_change() - df["Rfree"]
df["Excess_t+1"] = df["Excess"].shift(-1)
df.dropna(axis=0, inplace=True)
df.tail(5)
```
